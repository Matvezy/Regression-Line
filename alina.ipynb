```
import pandas as pd
```

```
data = pd.read_csv("https://filtereddatasets.s3.amazonaws.com/Groceries_retail/Groceriesv2.csv")
```

```
df = pd.DataFrame(data)
df.head()
```

# Problem #1

```
prod_names = df['Prod_name'].unique()
appearences = {key: {key1:0 for key1 in [i for i in prod_names if i != key]} for key in prod_names}
appearences['Icecream Cone - Areo Chocolate']
```

```
sep_ord = df.groupby(["Order_no", "Prod_name"])["Prod_name"].count()
ord_nums = df['Order_no'].unique()
```

```
for num in ord_nums:
  for product in sep_ord[num].keys():
    for ad_product in sep_ord[num].keys():
      if ad_product != product:
        appearences[ad_product][product] += 1
```

```
appearences['Icecream Cone - Areo Chocolate']
```

```
max = 0
product = ''
product1 = ''
for key in appearences.keys():
  for key1 in appearences[key].keys():
    if appearences[key][key1] > max:
      product = key
      product1 = key1
      max = appearences[key][key1]
```

```
print("The products that are purchased together at the greatest number of orders are: '" + product + "' and '" + product1 + "'")
print("Number of times purchased together: " + str(max))
```

# The products that are purchased together at the greatest number of orders are: 'Icecream Cone - Areo Chocolate' and 'Icecream - Dstk Cml And Fdg'
# Number of times purchased together: 300
